```python
# these values are in millions
# lets find the movie with the highest budget
highest_budget = movies["budget_musd"].dropna().max()
movies[movies["budget_musd"] == highest_budget]
```

```python
# lets find the movie with the highest budget
lowest_budget = movies["budget_musd"].dropna().min()
lowest_budget_movies = movies[movies["budget_musd"] == lowest_budget]
```

```python
converted_to_datetime = list(pd.to_datetime(lowest_budget_movies["release_date"]))
index_of_low_budget_movies = list(lowest_budget_movies.index)
```

```python
index_pos = 0
for index in index_of_low_budget_movies:
    lowest_budget_movies.loc[index, "release_date"] = converted_to_datetime[index_pos]
    index_pos += 1
```

```python
list(converted_to_datetime.values)[0]
```

```python
lowest_budget_movies.info()
```

```python
# quite a few movies are released with the same lowest budget let's find the most recent one
lowest_budget_movies.sort_values("release_date", ascending=False).head(1)
```

```python
max_revenue = movies["revenue_musd"].max()
max_revenue_movie = movies[movies["revenue_musd"] == max_revenue]
HTML(max_revenue_movie.to_html(escape= False))
#avatar had the highest revenue
```

```python
# chaaging the release_date column to datetime column for the entire dataframe
movies["release_date"] = pd.to_datetime(arg=movies["release_date"])
```

```python
movies.info()
```

```python
# best movie in terms of revenus
min_revenue = movies["revenue_musd"].min()

#multiple movie had the same minimum revenue lets find the most recent one
movies[movies["revenue_musd"] == min_revenue].sort_values("release_date", ascending=False).head(1)
```

```python
# Proft, Ratings ..
# quite a lot of movies with an average rating of 10
highest_rated_movies = movies[movies["vote_average"] == movies["vote_average"].max()]
HTML(highest_rated_movies.to_html(escape= False))
```

```python
#lets find the movies with the lowest ratings
# quite a lot of movies with very low user ratings
movies[movies["vote_average"] == movies["vote_average"].min()]
```

```python
# Profit
# there is no profit column so let's add a profit column to the end of our dataframe and find out the most profit 
# making and the most loss making movies
movies.info()
```

```python
profit = movies["revenue_musd"] - movies["budget_musd"]
```

```python
movies["profit"] = profit
```

```python
# most profitable movie
movies[movies["profit"] == movies["profit"].max()]
```

```python
# lets find the least profitable movie
# this movie made a loss
movies[movies["profit"] == movies["profit"].min()]
```

```python
budget_to_profit_ratio = movies["budget_musd"] / movies["profit"]
```
