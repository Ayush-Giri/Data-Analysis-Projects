```python
import requests
import pandas as pd
url = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/dailyActivity_merged.csv'
res = requests.get(url, allow_redirects=True)
with open('smart_watch_data.csv','wb') as file:
    file.write(res.content)
smartwatch_date = pd.read_csv('smart_watch_data.csv')
```

```python
smartwatch_date
```

```python
# we have all the data we need i.e no column has null values
smartwatch_date.info()
```

```python
# here we are converting the activity date column to datetime column
smartwatch_date["ActivityDate"] = pd.to_datetime(smartwatch_date["ActivityDate"])
```

```python
smartwatch_date.info()
```

```python
smartwatch_date
```

```python
smartwatch_date["total_minutes"] = smartwatch_date["VeryActiveMinutes"] + smartwatch_date["FairlyActiveMinutes"] + smartwatch_date["LightlyActiveMinutes"] + smartwatch_date["SedentaryMinutes"]
```

```python
cropped = smartwatch_date[["TotalSteps", "Calories"]].sort_values("TotalSteps")
cropped
```

```python
import matplotlib.pyplot as plt
plt.scatter(cropped['TotalSteps'], cropped['Calories'])
plt.xlabel("steps")
plt.ylabel("calories")
plt.show()
# as we can number of steps directly does not co-relate to number of calories burned
# i.e more number of steps does not mean more number of calories burned various other factors also play a major role
```

```python
smartwatch_date[["VeryActiveMinutes", "LightlyActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes"]].mean().plot(kind="pie")
```

```python
smartwatch_date["Day"] = smartwatch_date["ActivityDate"].dt.day_name()
```

```python
smartwatch_date
```

```python
# lets find out on which day the use burns the most colorie
# the most active for the user throughout the dataset is tuesday
smartwatch_date[["VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", "Day", "Calories"]].groupby("Day")["Calories"].sum().sort_values().plot(kind="bar")
```
